```python
import pandas as pd
```

```python
chekcpoints = [0,5,10,12,15,30,50]

dataframes = [pd.read_csv(f"PredictedRaces_ckpt{i}.csv") for i in chekcpoints]
```

```python
dataframes[0]
```

```python
import matplotlib.pyplot as plt

# Create a figure and axes for subplots
fig, axes = plt.subplots(1, len(chekcpoints), figsize=(20, 5), sharey=True)

# Plot histograms for each checkpoint
for i, df in enumerate(dataframes):
    df['race'].value_counts().plot(kind='bar', ax=axes[i], title=f'Checkpoint {chekcpoints[i]}')
    axes[i].set_xlabel('Race')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()

```
